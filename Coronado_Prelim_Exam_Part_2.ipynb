**Instructions:**

You have an image named circles.png containing various circles  circles Links to an external site. of different sizes. Write a Python script using OpenCV that:

    Load and preprocess the image
    Detect circles using Hough Circle Transform
    Draw detected circles in red color
    Save the result as circles_detected.png




```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
#Load and preprocess the image
img = cv2.imread("/content/drive/MyDrive/CSV/circles.png")
cv2_imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = cv2.GaussianBlur(gray, (9, 9), 2)

circles = cv2.HoughCircles(
    gray,
    cv2.HOUGH_GRADIENT,
    dp=1.1,
    minDist=140,
    param1=180,
    param2=45,
    minRadius=20,
    maxRadius=350
)

if circles is not None:
    circles = np.uint16(np.around(circles))
    print("Detected circles:", len(circles[0]))

# to make circles red
    for (x, y, r) in circles[0]:
        cv2.circle(img, (x, y), r, (0, 0, 255), 4)
        cv2.circle(img, (x, y), 2, (0, 0, 255), 3)

cv2.imwrite("circles_detected.png", img)
cv2_imshow(img)
print("Circle detection complete. Saved as circles_detected.png")
```
